```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("mn.csv")
df
```

```python
kolom = pd.read_csv("mn_headers_updated.csv")
kolom
```

```python
#merubah nama kolom sesuai dengan nama kolom asli
mn_header_label = dict(zip(kolom['Name'], kolom['Label']))
df.rename(columns=mn_header_label,inplace=True)
df
```

```python
#cek nama kolom
df.columns
```

```python
#cek duplikat berdasarkan kolom household number
df.duplicated(subset='Household number').sum()
```

```python
df = df.loc[:, ~df.columns.duplicated()]
```

buat cek atau remove kolom yang double

```python
df.drop(columns=['wscoreu', 'windex5r', 'mwelevel', 'mnweight', 'wscorer', 'windex5u'], inplace=True)
```

drop kolom yang gapunya nama

```python
df.duplicated(subset=['Cluster number','Household number']).sum()
```

```python
# Menghapus baris yang duplikat berdasarkan kombinasi kolom 'Cluster number' dan 'Household number'
df.drop_duplicates(subset=['Cluster number', 'Household number'], inplace=True)
```

```python
#cek ulang duplikasi
df.duplicated(subset=['Cluster number','Household number']).sum()
```

```python
df.dtypes
```

```python
#merubah semua tipe data float menjadi int
for col in df.select_dtypes(include=['float']).columns:
    df[col] = df[col].fillna(0).replace([float('inf'), float('-inf')], 0).round().astype(int)
```

```python
#cek outlier
# Menentukan kolom numerik
numeric_cols = df.select_dtypes(include=['number']).columns

# Mendeteksi outlier
outliers = {}
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Menyimpan indeks baris dengan outlier
    outliers[col] = df[(df[col] < lower_bound) | (df[col] > upper_bound)].index.tolist()

# Menampilkan hasil
for col, idx in outliers.items():
    print(f"Outliers in {col}: {idx}")
```

```python
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Ganti outlier dengan modus
    mode_value = df[col].mode()
    df.loc[(df[col] < lower_bound) | (df[col] > upper_bound), col] = mode_value
```

```python
# Membuat boxplot untuk semua kolom numerik
for col in numeric_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot for {col}')
    plt.show()
```
