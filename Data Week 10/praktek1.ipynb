```python
import pandas as pd
```

```python
#sample data
data = {'Product':['A','B','A','C','B','C'],
        'Category':['Electronics','Electronics','Furniture','Furniture','Furniture','Electronics'],
        'Sales':[200,300,150,450,250,500]}

df = pd.DataFrame(data)

#groupby
category_sales = df.groupby('Category')['Sales'].sum()
print(category_sales)
```

```python
#create new column with sales category based on value
df['Sales_Category'] = pd.cut(df['Sales'],bins=[0,200,400,600],
                              labels=['Low','High','Medium'])

#group by sales category
sales_grouped = df.groupby('Sales_Category')['Sales'].mean()
print(sales_grouped)
```
